```python
def get_data(url, params):
    print(params)
    response = requests.get(url, params=params)
    data = response.json()
    print(data)
    if 'Data' in data and 'Data' in data['Data']:
        return pd.DataFrame(data['Data']['Data'])
    elif 'Data' in data:
        return pd.DataFrame(data['Data'])
    else:
        print("Error fetching data")
        return None
```

```python
import pandas as pd
import xmltodict

def unique_code_xml_to_dataframe(xml_file):
    # XML 파일을 읽기
    with open(xml_file, 'r', encoding='utf-8') as file:
        xml_content = file.read()
    
    # XML을 파이썬 딕셔너리로 변환
    xml_dict = xmltodict.parse(xml_content)
    
    # 'result' 안의 'list'를 추출
    records = xml_dict['result']['list']
    
    # DataFrame 생성
    df = pd.DataFrame(records)
    
    return df
```

```python

import pandas as pd
import FinanceDataReader as fdr
import exchange_calendars as ecals


def get_active_date(date):
    XKRX = ecals.get_calendar("XKRX") # 한국 코드
    if XKRX.is_session(date):
        return pd.to_datetime(date).strftime('%Y-%m-%d')
    else:
        return pd.to_datetime(XKRX.next_open(date)).strftime('%Y-%m-%d')

price_df = pd.DataFrame()

for index, row in merged_full_data_X.iterrows():
    code = row['종목코드']
    start_date = row['공시일']
    end_date_1month = get_active_date(start_date + pd.DateOffset(months=1))
    end_date_3month = get_active_date(start_date + pd.DateOffset(months=3))
    end_date_6month = get_active_date(start_date + pd.DateOffset(months=6))
    # print(end_date_6month.strftime('%Y-%m-%d'))
    price = fdr.DataReader(code, start_date, end_date_6month)['Close']
    # print(price)
    price_1month = price.loc[price.index == end_date_1month]
    price_3month = price.loc[price.index == end_date_3month]
    price_6month = price.loc[price.index == end_date_6month]
    
    # price_1month = price.iloc[str(end_data_1month)]
    # price_3month = price.loc[price['Date'] ==end_data_3month]
    # price_6month = price.loc[price['Date'] ==end_data_6month]
    df_list = [price_1month, price_3month, price_6month]
    df = pd.concat(df_list, axis=0)
    df = df.reset_index().T.reset_index().drop(columns='index').drop(index=0)
    new_columns = {
        0: 'price_1month',
        1: 'price_3month',
        2: 'price_6month'
    }
    df = df.rename(columns=new_columns).reset_index(drop=True)
    df['종목코드'] = code
    df['공시일'] = start_date
    # 인덱스를 제거하고 열 이름을 설정
    price_df = pd.concat([price_df, df], ignore_index=True)
    print(index)

price_df
     
```
