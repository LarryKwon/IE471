```python
def get_data(url, params):
    print(params)
    response = requests.get(url, params=params)
    data = response.json()
    print(data)
    if 'Data' in data and 'Data' in data['Data']:
        return pd.DataFrame(data['Data']['Data'])
    elif 'Data' in data:
        return pd.DataFrame(data['Data'])
    else:
        print("Error fetching data")
        return None
```

```python
import pandas as pd
import xmltodict

def unique_code_xml_to_dataframe(xml_file):
    # XML 파일을 읽기
    with open(xml_file, 'r', encoding='utf-8') as file:
        xml_content = file.read()
    
    # XML을 파이썬 딕셔너리로 변환
    xml_dict = xmltodict.parse(xml_content)
    
    # 'result' 안의 'list'를 추출
    records = xml_dict['result']['list']
    
    # DataFrame 생성
    df = pd.DataFrame(records)
    
    return df
```

```python
def concat_unique_code(company_df, unique_code_df):
    company_df = company_df.merge(unique_code_df, left_on='종목명', right_on='corp_name', how='left')
    return company_df
```

```python
def concat_unique_code_by_stock_code(company_df, unique_code_df):
    company_df = company_df.merge(unique_code_df, left_on='종목코드', right_on='stock_code', how='left')
    return company_df
```
